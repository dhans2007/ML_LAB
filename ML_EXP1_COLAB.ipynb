IN-LAB TASKS:

4)Check for missing values.

```python
print("DHANESH VC 24BAD018")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print("DHANESH VC 24BAD018")
df = pd.read_csv("data.csv", encoding="latin1")
```

```python
print("DHANESH VC 24BAD018")
df['Sales'] = df['Quantity'] * df['UnitPrice']
product_sales = df.groupby('Description')['Sales'].sum().head(10)
product_sales.plot(kind='bar')
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.title("Top 10 Products by Sales")
plt.show()
```

```python
print("DHANESH VC 24BAD018")
product_sales.plot(kind='line')
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.title("Sales Trend of Top 10 Products")
plt.show()
```

Investigate a housing dataset for missing features and relationships between house attributes.

Dataset (Kaggle â€“ Public):

https://www.kaggle.com/datasets/harishkumardatalab/housing-price-prediction



IN-LAB TASKS:

1)Load the dataset and inspect columns.

2)Detect and summarize missing values.

3)Visualize correlations between numerical features using scatter plots and heatmaps.

4)Interpret trends for price prediction.

```python
print("DHANESH VC 24BAD018")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Housing.csv")
df.columns
```

```python
print("DHANESH VC 24BAD018")
df.isnull().sum()
```

```python
print("DHANESH VC 24BAD018")
plt.scatter(df['area'], df['price'])
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Area vs Price")
plt.show()
```

```python
print("DHANESH VC 24BAD018")

numeric_df = df.select_dtypes(include=['int64', 'float64'])

corr = numeric_df.corr()

sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
